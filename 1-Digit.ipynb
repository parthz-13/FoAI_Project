```python
def leading_digit(x):
    try:
        x = abs(float(x))


        cleaned = re.sub(r'[^\d]', '', f"{x:.15g}")


        return int(cleaned[0]) if cleaned else None
    except:
        return None

def Galatic_long(df):
    df["Galactic Longitude Leading"] = df["Galactic Longitude"].apply(leading_digit)
    df = df.dropna(subset=["Galactic Longitude Leading"])

    proportions = df["Galactic Longitude Leading"].value_counts(normalize=True).sort_index()
    return proportions
y = Galatic_long(df).drop(Galatic_long(df).index[0])
plt.bar(y.index,y.values,color="lightpink")
plt.title("Leading Digit Frequency")
plt.xticks(y.index)
plt.ylabel("Frequency")
plt.xlabel("Leading Digit")

plt.show()

```

```python
def Galatic_lat(df):
    df["Galactic Latitude Leading"] = df["Galactic Latitude"].apply(leading_digit)
    df = df.dropna(subset=["Galactic Latitude Leading"])

    proportions = df["Galactic Latitude Leading"].value_counts(normalize=True).sort_index()

    return proportions

x= Galatic_lat(df).drop(Galatic_lat(df).index[0])
plt.bar(x.index,x.values,color="peachpuff")
plt.title("Leading Digit Frequency")
plt.xticks(x.index)
plt.xlabel("Leading Digit")
plt.ylabel("Proportion")
plt.show()
```

```python
def distance(df):
    df["Distance (ly) Leading"] = df["Distance (ly)"].apply(leading_digit)
    df = df.dropna(subset=["Distance (ly) Leading"])

    proportions = df["Distance (ly) Leading"].value_counts(normalize=True).sort_index()

    return proportions
x=distance(df)
plt.bar(x.index,x.values,color="lightcoral")
plt.title("Leading Digit Frequency")
plt.xticks(x.index)
plt.xlabel("Leading Digit")
plt.ylabel("Proportion")
plt.show()
```

```python
y = pd.concat([Galatic_long(df),
               Galatic_lat(df),
               Visual_Magi(df),
               Absolute_Magi(df),
               Parallax(df),
               Error(df),
               distance(df)], axis=1).sum(axis=1, skipna=True)

y=y/7
y = y.drop(y.index[0])

plt.bar(y.index, y.values, color='lightgreen', label='Bar')


plt.plot(y.index, y.values, color='blue', marker='o', linewidth=2, label='Line')
plt.title("Leading Digit Frequency")
plt.xticks(y.index)
plt.ylabel("Frequency")
plt.xlabel("Leading Digit")
plt.ylim(0,0.30)
plt.show()
```
